```python
again = 'yes'

while again == 'yes':
    menu = input('''
    If you want to add 2 numbers, enter +.

    If you want to subtract 2 numbers, enter -.

    If you want to multiply 2 numbers, enter x.

    If you want to multiply 2 numbers, enetr รท.

    Enter q to quit.

    ''')
    if menu == '+':
        num1 = int(input('''
        What is the first number you want to add?
        '''))
        num2 = int(input ('''What is the second number you want to add?
        '''))
    elif menu == '-':
        num1 = int(input('''
        What is the first number you want to subtract?
        '''))
        num2 = int(input ('''What is the second number you want to subtract?
        '''))
    elif menu == 'x':
        num1 = int(input('''
        What is the first number you want to multiply?
        '''))
        num2 = int(input ('''What is the second number you want to multiply?
        '''))
    elif menu == 'รท':
        num1 = int(input('''
        What is the first number you want to divide?
        '''))
        num2 = int(input ('''What is the second number you want to divide?
        '''))

    if menu == '+':
        answer = num1 + num2
        print(f'The answer is {answer}')
    elif menu == '-':
        answer = num1 - num2
        print(f'The answer is {answer}')
    elif menu == 'x':
        answer = num1 * num2
        print(f'The answer is {answer}')
    elif menu =='รท':
        answer = num1 / num2
        print(f'The answer is {answer}')
    elif menu == 'q':
        print('''
        exiting...''')
        break
    elif menu != '+' or menu != '-' or menu != 'x' or menu != 'รท' or menu != 'q':
        print('''
        Enter the correct symbol.
        ''')
    
    again = input('''
    Do you want to do another calculation? Enter yes or no.
    ''')
    if again == 'no':
        break
```
